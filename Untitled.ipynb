```python
from pptx import Presentation
from glob import glob
from PIL import Image
from pptx.util import Pt
```

```python
# Input
old_file_name = 'template'
new_file_name = 'new.pptx'
```

```python
# パワポの準備
ppt = Presentation('template.pptx')
width = ppt.slide_width
height = ppt.slide_height
```

```python
#使用するスライドの種類
title_slide_layout = ppt.slide_layouts[0] #Title Slideの作成
slide_1_layout = ppt.slide_layouts[1] # 写真1枚
slide_2_layout = ppt.slide_layouts[2] # 写真2枚
slide_3_layout = ppt.slide_layouts[3] # 写真3枚
```

```python
def replace_with_image(img, shape, slide):
    pic = slide.shapes.add_picture(img, shape.left, shape.top)

    #calculate max width/height for target size
    ratio = min(shape.width / float(pic.width), shape.height / float(pic.height))

    pic.height = int(pic.height * ratio)
    pic.width = int(pic.width * ratio)

    pic.left = shape.left + int((shape.width - pic.width) / 2)
    pic.top = shape.top + int((shape.height - pic.height) / 2)

    placeholder = shape.element
    placeholder.getparent().remove(placeholder)
    return
```

```python
#スライドに1枚画像貼り付け
slide_1 = ppt.slides.add_slide(slide_1_layout)
# for i in slide_1.placeholders:
#     print('%d %s' % (i.placeholder_format.idx, i.name))
placeholder_1 = [13]
fnms_1 = glob('img/*')
for i in range(0,1):
    fnm = fnms_1[i]
    placeholder = slide_1.placeholders[placeholder_1[i]]
    replace_with_image(fnm, placeholder,slide_1)
    
```

```python
#スライドに2枚画像貼り付け
slide_2 = ppt.slides.add_slide(slide_2_layout)
for i in slide_2.placeholders:
    print('%d %s' % (i.placeholder_format.idx, i.name))
placeholder_2 = [14,15]
fnms_2 = glob('img2/*')
for i in range(0,2):
    fnm = fnms_2[i]
    placeholder = slide_2.placeholders[placeholder_2[i]]
    replace_with_image(fnm, placeholder,slide_2)
```

```python
#スライドに3枚画像貼り付け
slide_3 = ppt.slides.add_slide(slide_3_layout)
for i in slide_3.placeholders:
    print('%d %s' % (i.placeholder_format.idx, i.name))
placeholder_3 = [14,15,16]
fnms_3 = glob('img3/*')
for i in range(0,3):
    print(fnm)
    fnm = fnms_3[i]
    placeholder = slide_3.placeholders[placeholder_3[i]]
    replace_with_image(fnm, placeholder,slide_3)
```
